```python
#SKlearn SVM untuk classification
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
import pandas as pd

df = pd.read_csv('diabetes.csv')
#memisahkan atribut pada dataset dan menyimpannya pada sebuah variabel
x = df[df.columns[:8]]
#memisahkan label pada dataset dan menyimpannya pada sebuah variable
y = df['Outcome']
df.head()
#melakukan standarisiasi nilai pada dataset
scaler = StandardScaler()
scaler.fit(x)
x = scaler.transform(x)
#memisahkan train data dan testing
x_train,x_test,y_train,y_test = train_test_split(
    x,y,test_size= 0.33,
    random_state = 42
)

clf = SVC()
clf.fit(x_train,y_train)
clf.score(x_test,y_test)

tree_model = DecisionTreeClassifier()

tree_model.fit(x_train , y_train)

print(tree_model.score(x_test,y_test))
print(tree_model.predict([[1000,2300,0,1,5,6.78,500,11899]]))
```
